Matplotlib :

 Useful for making plots.

```python
# Importing matlotlib library
import matplotlib.pyplot as plt
```

```python
# importing numpy to take random values for plots
import numpy as np
```

```python
x=np.linspace(0,10,100) # 100 points n range of 0 to 100
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting data

```python
# sin wave
plt.plot(x,y)
plt.show()
```

```python
# cosine wave
plt.plot(x,z)
plt.show()
```

```python
# adding title , x-axis,y-axis labels
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sin value')
plt.title('sin wave')
plt.show()
```

```python
# Parabola
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

Bar chart

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])  # 0,0 for origin & 1,1 for height & width
train=['train data','test data']
values=[80,20]
ax.bar(train,values)
plt.xlabel('data')
plt.ylabel('values')
plt.title('bar chart')
plt.show()
```

Pie chart

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['English','German','Latin','French']
values=[100,60,50,70]
ax.pie(values,labels=languages,autopct='%1.1f%%')
plt.show()
```

Scatter Plot

```python
x=np.linspace(5,10,25)
y=np.sin(x)
z=np.cos(x)
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='r')
plt.show()
```

3-D scatter plot

```python
fig=plt.figure()
ax=plt.axes(projection='3d')
a=20*np.random.randn(100)
b=np.sin(a)
c1=np.cos(a)
ax.scatter(b,c1,a,c=a,cmap='Blues')
plt.show()
```

Histogram

```python
data=[20,30,40,50,50,70]
plt.hist(data,bins=4,color='skyblue',edgecolor='black')
plt.title('Histogram')
plt.show()
```

Saving plot as image:

```python
plt.hist(data,bins=3,color='skyblue',edgecolor='black')
plt.title('Histogram2')
plt.savefig('histogram.png')
plt.show()
```
