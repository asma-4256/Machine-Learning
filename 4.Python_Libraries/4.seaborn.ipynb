Seaborn :

 Data visualization library

Importing libraries :

```python
import seaborn as sns
import matplotlib as plt
import pandas as pd
import numpy as np
```

Seaborn has some built in datasets [load_dataset() method]

```python
# loading the sample dataset about restuarant bills :
data=sns.load_dataset('tips')
```

```python
data.head()
```

```python
# settng the theme for plots
sns.set_theme()
#data.shape
```

```python
#visualizing the data
sns.relplot(x='total_bill',y='tip',data=data)
```

```python
# visualizing data with one factor
sns.relplot(x='total_bill',y='tip',data=data,hue='smoker',style='smoker',size='size',col='time')
```

Scatter plot

```python
sns.scatterplot(x='total_bill',y='tip',data=data,hue='smoker')
```

```python
sns.scatterplot(x='total_bill',y='tip',data=data,hue='smoker',style='smoker',size='size')
```

```python
#line plot
sns.lineplot(x='total_bill',y='tip',hue='smoker',data=data)
```

```python
# loading the house dataset
from sklearn.datasets import fetch_california_housing
house_=fetch_california_housing()
house=pd.DataFrame(house_.data,columns=house_.feature_names)
house['prize']=house_.target
```

```python
house.head()
```

Distribution plot

```python
sns.distplot(house['prize'])
```

Correlation :

1.postive correlaton

2.negative correlation

Heat Map

```python
correlation=house.corr()
#plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues') # light color represents -ve correlaton & dar
```
