Addition of Vectors

```python
vec1=np.asarray([0,0,2,3])
vec2=np.asarray([0,0,3,-2])
sum=vec1+vec2
print(sum)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='red')
plt.quiver(0,0,3,-2, scale_units='xy', angles='xy', scale=1, color='y')
plt.quiver(0,0,5,1, scale_units='xy', angles='xy', scale=1, color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Subtraction of two vectors

```python
vec1=np.asarray([0,0,2,3])
vec2=np.asarray([0,0,3,-2])
sub=vec1-vec2
print(sub)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='red')
plt.quiver(0,0,3,-2, scale_units='xy', angles='xy', scale=1, color='y')
plt.quiver(0,0,-1,5, scale_units='xy', angles='xy', scale=1, color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Cross product of two vectors -vector

```python
a=np.array([2,4])

b=np.array([2,3])

a_cross_b=np.cross(a,b)

print(a_cross_b)

```

```python
a=np.array([30,34,36])

b=np.array([40,43,46])

a_cross_b=np.cross(a,b)

print(a_cross_b)
```

Projection of "a" vector on "b" vector

pro =(a.b  / ( ||b|| ^2))*b

```python
a=np.array([2,4])

b=np.array([2,3])
# magntude of b
mag_b=np.sqrt(sum(b**2))

a_dot_b=np.dot(a,b)

proj_a_on_b=(a_dot_b/mag_b**2)*b

print(proj_a_on_b)
```
