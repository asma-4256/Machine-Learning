Imbalanced Dataset:
* The number of observations in one class is higher than in another class

Challenges :
* Model may ignore the minority class by only focusing on majority class
* Accuracy may low (misleading)
* Hard to detect patterns in minority classes

**Handling Imbalanced Datasets**

```python
# importing the libraries
import pandas as pd
```

```python
credit_data=pd.read_csv('/content/credit_data.csv')
credit_data.head()
```

```python
credit_data.tail()
```

```python
credit_data.shape
```

```python
credit_data['Class'].value_counts()
```

Imbalanced dataset:

0--> Legit transactions
1--> Fradulant transactions

```python
#Seperating the Legit & Fradulant transactions classes
legit=credit_data[credit_data.Class==0]
fraud=credit_data[credit_data.Class==1]
```

```python
print(legit.shape)
fraud.shape
```

Resampling techniques(data level)
* Oversampling : Increases the instances of minority class ,SMOTE - Synthetic Minority Over-sampling Technique).
* Undersampling : Decreases the instances of majority class

Under sampling
* Build a sample dataset containing similar distribution of Legit & Fraudulent transactions
* Number of Fraudulent transactions --> 492


```python
#Sampling the majority class to 492 instances
legit_sample=legit.sample(n=492)
legit_sample.shape
```

```python
# Concatinate the two dataframes
credit_data_new=pd.concat([legit_sample,fraud],axis=0)
```
