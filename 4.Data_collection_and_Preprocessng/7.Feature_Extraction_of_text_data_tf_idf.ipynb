Feature Extraction of text data :
* Extracting features from text data

**Methods**

1. Bag of words
2. TF-IDF(Term Frequency-Inverse Document Frequency)

TF-IDF method :
* Text vectorization technique that converts textual data into numerical features for machine learning models

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
news=pd.read_csv('/content/train.csv')
```

```python
news.shape
```

```python
news.head()
```

```python
# filling missing values
news['text'] = news['text'].fillna('')
```

```python
print(news['text'])
```

Tf-idf Vectorizer

```python
#convert the textual data to feature vectors
vectorizer=TfidfVectorizer()
```

```python
vectorizer.fit(news['text'])
```

```python
X=vectorizer.transform(news['text'])
```

```python
print(X)
```
