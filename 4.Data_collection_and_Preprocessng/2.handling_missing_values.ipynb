Imputation
* Replacing missing values with statstical measures

Statistical measures(central tendency)
* mean
* median
* mode

```python
#loading data
df=sns.load_dataset('titanic')
```

```python
df.isnull().sum()
```

```python
# analysing the distribution of data
fig, ax = plt.subplots(figsize=(8,8))
#sns.distplot(df.deck)
sns.countplot(x='deck', data=df)
```

```python
df['deck'].fillna(df['deck'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

Interpolaton

Replacing missing values with neighbouring data points

```python
df['age'] = df['age'].interpolate(method='linear')
```
